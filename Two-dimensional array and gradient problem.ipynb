# Problem 1:
Linear Function

```python
import numpy as np
np.set_printoptions(suppress = True) # remove the scientific notation
x = np.arange(-50,50.1,0.1)
print(x)
y = np.array([])
for num in x:
    y = np.append(y, ((num/2)+1))
print(y)
print(y.shape)
```

# Problem 3:
Find the gradient: 
Find gradient at each x-coordinates 
gradient = amount of change in Y / amount of change in X


```python
gradient_list = []
for i in range(1001):
    gradient_list.append(((a[i,1] - a[(i-1),1]) / (a[i,0] - a[(i-1),0])))
print(gradient_list)
print(len(gradient_list))
    
```

# Problem 5: 
Python Functionalization

```python
#1. gradient of equation y=x^2 range(-50,0.1,50)
def compute_gradient(function, x_range):
    xrange = np.arange(*x_range)# *unpacks the elements of the tuple and passes them as individual arguments to the function
    yrange = np.array([])
    for _ in xrange:
        yrange = np.append(yrange,function)
        if len(yrange) == len(xrange):
            break
    a = np.column_stack((xrange,yrange))
    gradient = np.array([])
    for i in range(1,len(a[:,0])):# avoids the gradient of oth index which would be zero
        gradient=np.append(gradient, ((a[i,1] - a[(i-1),1]) / (a[i,0] - a[(i-1),0])))
    print('The gradient is {}'.format(gradient))

compute_gradient((pow(x,2)), (-50,50.1,0.1))   
```
